```python
from sklearn.datasets import make_regression
import numpy as np
from matplotlib import pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

# Prepare data

```python
x, y = make_regression(n_samples=100, n_features=1, noise=25.0, random_state=0)

# Scale x to 0..5 range
x = np.interp(x, (x.min(), x.max()), (0, 5))

# Scale y to 0..100 range
y = np.interp(y, (y.min(), y.max()), (0, 100))
```

```python
plt.scatter(x, y)
plt.title('One feature dataset')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Sklearn Linear Regression

```python
reg = LinearRegression().fit(x, y)

y_pred = reg.predict(x)

# The coefficients
print('Coefficients: \n', reg.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y, y_pred))
```

```python
# Plot outputs
plt.scatter(x, y)
plt.plot(x, y_pred, color='red', linewidth=3)

plt.title('Sklearn Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Mean Square Error implementation

```python
def mean_squared_error_local(first, second):
    length = len(first)
    if length != len(second):
        raise ValueError(f'Found input variables with inconsistent numbers of samples: [{len(first)}, {len(second)}]')
    
    sum = 0
    for i in range(0, length):
        sum += (first[i] - second[i]) ** 2
    
    return sum / length
```

```python
x_1 = np.arange(0., 5., 0.05)
f_1 = 36 * x_1 + 1

plt.scatter(x, y)
plt.plot(f_1, color='red', linewidth=3)

plt.xlim(-0.5, 5.5)
plt.ylim(-10, 110)

plt.title('First iteration')
plt.xlabel('t')
plt.ylabel('f_1')
plt.show()
```

```python
print(f'Sklearn MSE: {mean_squared_error(y, f_1)}')
print(f'Numpy MSE: {np.square(np.subtract(y, f_1)).mean()}')
print(f'Local MSE: {mean_squared_error_local(y, f_1)}')
```
